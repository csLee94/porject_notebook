```
df['SalePrice'].describe()
```

```
plt.figure(figsize=(16,9))
sns.distplot(x=df['SalePrice'], color='green')
```

```
df_25 = df.loc[df['SalePrice']<=129975]
df_75 = df.loc[df['SalePrice']>=214000]
df_25['label']='25%'
df_75['label']='75%'
```

```
df2575= pd.concat([df_25, df_75])
df2575
```

```
tmplst = list(df2575.columns)
tmplst.remove('Id')
tmplst.remove('SalePrice')
tmplst.remove('label')
for tmptxt in tmplst:
  if type(df2575[tmptxt].iloc[0]) == str:
    fig, ax = plt. subplot(8,5)
    
    sns.displot(data= df2575, x='MSSubClass', hue='label').set_title(tmptxt)


# for tmptxt in 
# sns.displot(data= df2575, x='MSSubClass', hue='label')
```
