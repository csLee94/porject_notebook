# Target 변수 탐색
- **Dwelling & Lot에 관한 여러가지 정보와 그에 대한 Price of Sale**
1. SalePrice 변수를 기준으로 상위 25%, 하위 25% 그룹 별 비교를 통해 변수별 차이 EDA

```
df['SalePrice'].describe()
```

```
plt.figure(figsize=(12,6))
sns.boxplot(df['SalePrice'], color = 'green', showmeans=True)
plt.show()
```

```
print('왜도: ', df['SalePrice'].skew())
print('첨도: ', df['SalePrice'].kurt())
```

```
plt.figure(figsize=(16,9))
sns.distplot(x=df['SalePrice'], color='green')
```

```
# outlier 제거를 위한 iqr
iqr = df['SalePrice'].describe()['75%'] - df['SalePrice'].describe()['25%']
iqr
```

```
df2 = df.loc[df['SalePrice']<= iqr*4]
print(df2['SalePrice'].describe())

print("왜도: ",df2['SalePrice'].skew())
print("첨도: "df2['SalePrice'].kurt())
```

```
plt.figure(figsize=(16,9))
sns.distplot(x=df2['SalePrice'], color = 'green')
plt.show()
```

```
sns.boxplot(df2["SalePrice"], color='green')
```

```
#outlier 그룹 분리
df_out = df.loc[df['SalePrice']>4*iqr]
df_out
```

```
# df['Label'] = 'temp'
df['Label'].loc[df['SalePrice']<=df['SalePrice'].describe()['25%']] = '25%'
df['Label'][df['SalePrice'].between(df['SalePrice'].describe()['75%'],4*iqr)] = '75%'
df['Label'].loc[df['SalePrice']>4*iqr] = 'Outlier'

df
```

```
df["Label"].value_counts()
```

```
df.loc[df['SalePrice']<=df['SalePrice'].describe()['25%']]
```

```
# 75% 이상 outlier 미만 그룹
mask = (df['SalePrice']>=df['SalePrice'].describe()['75%']) & (df["SalePrice"]<= 4*iqr)
df_75 = df[mask]

# 0 이상 25% 미만 그룹
df_25 = df.loc[df['SalePrice']<=df['SalePrice'].describe()['25%']]
```

```
fdf = pd.concat([df_25, df_75, df_out])
fdf
```
